```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
experiment_path = "../results/all_db_all_training"
```

```python
full_result_df = pd.DataFrame(columns=["experiment", "dataset", "f1", "auroc", "training_time", "inference_time"])
```

```python
for experiment_name in os.listdir(experiment_path):
    for dataset_name in os.listdir(experiment_path + "/" + experiment_name):
        try:
            print(experiment_path + "/" + experiment_name + "/" + dataset_name + "/results.csv")
            result_df = pd.read_csv(experiment_path + "/" + experiment_name + "/" + dataset_name + "/model_metrics.csv")
        except:
            continue
        print('Result parsed')
        result_df["experiment"] = experiment_name
        result_df["dataset"] = dataset_name
        full_result_df = pd.concat([full_result_df, result_df])
```
