```python
# Perform a TSNE on the data
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
X_embedded = tsne.fit_transform(X)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.scatter(X_embedded[y == 0, 0], X_embedded[y == 0, 1], c='b', label='0')
plt.scatter(X_embedded[y == 1, 0], X_embedded[y == 1, 1], c='r', label='1')
plt.legend()
plt.show()
```
